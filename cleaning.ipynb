```python
import pandas as pd
df = pd.read_csv("cars_24_combined.csv")
df.head ()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.drop(columns=["Unnamed: 0"])
df.head ()
```

```python
df.isnull().sum()
```

```python
df[df["Car Name"].isnull()]
```

```python
df = df.dropna(subset=["Car Name", "Year"])
df.shape
```

```python
df["Year"].dtype
```

```python
df["Year"] = df["Year"].astype("int64")
df["Year"].dtype
```

```python
df.duplicated().sum()
```

```python
df["Location"].isnull().sum()
```

```python
df["Location"] = df["Location"].fillna("unknown")
```

```python
df["Location"].isnull().sum()
```

```python
df["Car Name"].head()
```

```python
df["Car Name"].str.lower().head()
```

```python
df["Car Name"].nunique( )
```

```python
df["Car Name"].str.split().head()
```

```python
df["Brand"] = df["Car Name"].str.split().str[0]
df["Model"] = df["Car Name"].str.split().str[1:].str.join(" ")
df[["Car Name", "Brand", "Model"]].head()
```

```python
df["Brand"] = df["Brand"].str.title()
df["Model"] = df["Model"].str.title()
df[["Brand", "Model"]].head()
```

```python
df[["Car Name", "Brand", "Model"]].sample(10)
```

```python
df = df.drop(columns=["Car Name"])
df.head()
```

```python
df["Fuel"].unique()
```

```python
df["Fuel"] = df["Fuel"].str.title()
df["Fuel"].unique()
```

```python
df["Fuel"] = df["Fuel"].replace({
    "Cng" : "CNG",
    "Lpg" : "LPG"
})
```

```python
df["Drive"].unique()
```

```python
df["Type"].unique()
```

```python
df["Type"] = df["Type"].replace({
    "HatchBack" : "Hatchback",
    "Lux_SUV" : "Lux SUV",
    "Lux_sedan" : "Lux Sedan"
})
df["Type"].unique()
```

```python
df.describe()
```

```python
df.to_csv("cars24_cleaned.csv", index=False)
```
